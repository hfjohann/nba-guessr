```python
import os
import pandas as pd
from tabulate import tabulate
input_data = pd.read_excel(os.path.join("data","stats.xlsx"), header=1)
```

```python
input_data.columns = ['RANK', 'NAME', 'TEAM', 'POS', 'AGE', 'GP', 'MPG',
       'MIN%','USG%', 'TO%', 'FTA', 'FT%', '2PA', '2P%', '3PA', '3P%',
       'eFG%', 'TS%', 'PPG', 'RPG','TRB%','APG','AST','SPG', 'BPG','TOPG',
       'VI', 'ORTG', 'DRTG']
input_data.drop(labels='RANK',axis=1, inplace=True)
```

```python
input_data.head()
```

```python
input_data.to_csv(os.path.join("data", "stats.csv"))
```

```python
p = 30
r = 10
a = 10
def f(df,_p=10,_r=10,_a=10):
    return df['PPG'].subtract(_p).pow(2) + df['RPG'].subtract(_r).pow(2) + df['APG'].subtract(_a).pow(2)
```

```python
p = 27
r = 7
a = 7
result = input_data.loc[f(input_data,p,r,a).sort_values().head(5).index]
print(tabulate(result[['NAME','TEAM','GP','PPG','RPG','APG']], headers='keys'))
```
